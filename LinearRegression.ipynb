```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
y_ = np.linspace(-5, 5, 100)
X_ = np.random.randn(100) + y_
```

```python
plt.scatter(X_, y_)
plt.grid()
plt.title("Sample of our data")
plt.show()
```

```python
X = np.mat(X_).T
y = np.mat(y_).T
```

```python
w_hat = ((X.T*X)**-1)*X.T*y
print("Calculated weights by stationarity", w_hat)
```

```python
plt.scatter(X_, y_)
plt.plot([-6, 6],[w_hat[0, 0]*-6, w_hat[0, 0]*6])
plt.grid()
plt.show()
```

```python
def grad(w):
    return -X.T*y+X.T*X*w
```

```python
wt = 30
step_size = 0.00015
```

```python
plt.scatter(X_, y_)
plt.xlim(-7, 7)
plt.ylim(-7, 7)

for i in range(1, 100):
    wt = wt - step_size*grad(wt)
    plt.plot([-6, 6],[wt[0, 0]*-6, wt[0, 0]*6])

plt.grid()
plt.show()    
```

```python
print(f"Gradient descend weights {wt} vs. solved weights {w_hat}.")
```
