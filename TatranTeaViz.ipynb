```python
import pandas as pd
import io
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
drinks_csv = """
name,color,percent,price,pmin,pmax
coconut,#eef1f2,22,419,247,457
citrus,#B4B125,32,459,297,492
peach,#E26B3F,42,499,349,533
original,#474644,52,539,348,605
forest fruit,#612961,62,579,422,699
outlaw,#776D62,72,619,464,666
"""

df = pd.read_csv(io.StringIO(drinks_csv), sep=",")
df = df.set_index("name")
```

```python
df["alcohol_ml"] = (df.percent / 100)*700
df["ml_for_czk"] = df.alcohol_ml / df.price
```

```python
names = df.index
labels = [str(percent) + "%\n" + str(price) + " Kƒç" for percent, price in zip(df.percent, df.price)]
colors = df.color.values
values = df.ml_for_czk.values
```

```python
fig, ax = plt.subplots()
domain = range(len(values))
plt.bar(domain, height=values, tick_label=names, color=colors)
ax.set_title("Tatratea best bang for the buck ‚Äì alcohol content for 1 Kƒç")
ax.set_ylabel("ml of ethanol/1 Kƒç")

percentages = df.percent.apply(lambda x: str(x)+"%")

for i, v in enumerate(labels):
    ax.text(i, .04, str(v), color='black', horizontalalignment="center")

plt.tight_layout()
plt.show()
```
